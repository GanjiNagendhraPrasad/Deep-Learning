```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('/content/Iris.csv')
df=df.drop(['Id'],axis=1)
df['Species'] = df['Species'].map({'Iris-setosa':0 , 'Iris-versicolor':1 , 'Iris-virginica':2}).astype(int)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
training_data = df.iloc[ : 140 , :]
validation_data = df.iloc[140 : 148 , :]
testing_data = df.iloc[148 : ]
```

```python
training_data.shape
```

```python
validation_data.shape
```

```python
testing_data.shape
```

```python
training_data_inde = training_data.iloc[: , :-1]
training_data_dep = training_data.iloc[: , -1]
```

```python
validation_data_inde = validation_data.iloc[: , :-1]
validation_data_dep = validation_data.iloc[: , -1]
```

```python
testing_data_inde = testing_data.iloc[: , :-1]
testing_data_dep = testing_data.iloc[: , -1]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.activations import relu,softmax
```

```python
training_data_inde.shape[1]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model = Sequential()

# input_layers to hiddel layer 1
model.add(Dense(units=10 ,
                kernel_initializer="he_uniform" ,
                activation="relu" ,
                input_dim = training_data_inde.shape[1])
)

# hiddel layer 1 to hidden layer 2
model.add(Dense(units=4 , kernel_initializer="he_uniform" ,activation="relu"))

# hiddel layer 2 to hidden layer 3
model.add(Dense(units=2 , kernel_initializer="he_uniform" ,activation="relu"))


# hiddel layer 3 to outputlayer
model.add(Dense(units= 3, kernel_initializer="glorot_uniform" ,activation="softmax")) ###### softmax
```

```python
model.compile(optimizer="adam" , loss = "categorical_crossentropy" , metrics=["Accuracy"])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
training_data_dep_converted = to_categorical(training_data_dep)
```

```python
training_data_dep_converted
```

```python
validation_data_dep_converted = to_categorical(validation_data_dep)
```

```python
# Training _process

model.fit(training_data_inde ,
          training_data_dep_converted ,
          batch_size=10 ,
          epochs=50,
          validation_data = (validation_data_inde,validation_data_dep_converted)
          )
```
