```python
training_data=df.iloc[:40000,:]
validation_data=df.iloc[40000:45000,:]
testing_data=df.iloc[45000:]
```

```python
training_data_inde=training_data.iloc[:,:-1]
training_data_dep=training_data.iloc[:,-1]
```

```python
validation_data_inde=validation_data.iloc[:,:-1]
validation_data_dep=validation_data.iloc[:,-1]
```

```python
testing_data_inde=testing_data.iloc[:,:-1]
testing_data_dep=testing_data.iloc[:,-1]
```

#Design the Binary ANN Architecure

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.activations import relu,softmax
```

```python
training_data_inde.shape[1]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model=Sequential()

# input_layers to hiddel layer 1
model.add(Dense(units=10,
                kernel_initializer='he_uniform',
                activation='relu',
                input_dim=training_data_inde.shape[1]))

# hiddel layer 1 to hidden layer 2
model.add(Dense(units=4,kernel_initializer='he_uniform',activation='relu'))

# hiddel layer 2 to hidden layer 3
model.add(Dense(units=2,kernel_initializer='he_uniform',activation='relu'))

# hiddel layer 3 to outputlayer
model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['Accuracy'])
```

```python
40000/1000
```

```python
model.summary()
```

```python
# Training _process

model.fit(training_data_inde,
          training_data_dep,
          batch_size=1000,
          epochs=5,
          validation_data=(validation_data_inde,validation_data_dep))
```
