#COLOR IMAGES

```python
f=plt.imread('/content/lena_color_512.tif')
```

```python
f.shape
```

```python
print(f)
```

```python
plt.imshow(f)  # NOT USE cmap='gray'  BCZ ITS COLOR IMAGE
plt.show()
```

```python
plt.figure(figsize=(5,5))

plt.subplot(2,2,1)
plt.title('R_channel')
plt.imshow(f[:,:,0],cmap='gray')   # all row/cols  and 0th channel

plt.subplot(2,2,2)
plt.title('G_channel')
plt.imshow(f[:,:,1],cmap='gray') # all row/cols  and 1th channel

plt.subplot(2,2,3)
plt.title('B_channel')
plt.imshow(f[:,:,2],cmap='gray')   # all row/cols  and 2th channel


plt.subplot(2,2,4)
plt.title('Complete_Image')
plt.imshow(f)

plt.show()
```
